```python
from mnist import MNIST
import numpy as np
import cupy as cp
import random
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def relu(x):
      return cp.where(x > 0, x, 0)
```

```python
def normalize(x):
    epsilon = 0.001
    u = cp.average(x, axis = 0)
    variance = cp.sum(cp.square(x-u), axis = 0, keepdims = True)
    return (x-u)/(cp.sqrt(variance + epsilon))
```

```python
def softmax(x): 
    ans = cp.exp(x) #-cp.max(x, axis=0, keepdims=True))
    d = cp.sum(ans, axis = 0)
    return ans/(d)
```

```python
def cost(Y, preds, m):
    res = np.zeros_like(preds)
    res[preds.argmax(0), np.arange(len(preds[0,:]))] = 1
    op = Y*cp.log(preds) 
    loss = -cp.sum(op,axis = 0, keepdims = True) 
    cost = (1/m)*cp.sum(loss, axis = 1) 
    return cost
```

```python
def initialize_weights():
    w1 = normalize(cp.random.rand(128, 784))
    b1 = cp.random.rand(128, 1)
    w2 = cp.random.rand(10,128)
    b2 = cp.random.rand(10, 1)
    normalize(w1)
    normalize(w2)
    return w1, w2, b1, b2
```

```python
def forward_prop(X, Y, w1, w2, b1, b2):
    z1 = cp.dot(w1, X) + b1
    #z1 = normalize_z(z1)
    a1 = relu(z1)
    z2 = cp.dot(w2, a1) + b2
    a2 = softmax(z2)
    return a1, a2
```

```python
def back_prop(X, Y, w1, w2, b1, b2, a1, a2, m):
    dz2 = a2-Y
    dw2 = (1/m)*cp.dot(dz2, a1.T)
    db2 = (1/m)*(cp.sum(dz2, axis = 1, keepdims = True))
    da1 = cp.dot(w2.T, dz2)
    dz1 = da1
    dz1[dz1 < 0] = 0
    dz1[dz1 > 0] = 0
    dw1 = (1/m)*cp.dot(dz1, X.T)
    db1 = (1/m)*(cp.sum(dz1, axis = 1, keepdims = True))
    return dw1, dw2, db1, db2
```

```python
def accuracy(X, Y, w1, w2, b1, b2):
    a1, pred = forward_prop(X, Y, w1, w2, b1, b2)
    res = np.zeros_like(pred)
    m = len(pred[0,:])
    res[pred.argmax(0), np.arange(len(pred[0,:]))] = 1
    #print(res[:,0], Y[:,0])
    unmatched = cp.count_nonzero(cp.sum(abs(Y-res), axis=0))
    return (1-(unmatched/m))*100, res, pred
```

```python
w1, w2, b1, b2 = initialize_weights()
```

```python
m = 1000
alpha = 0.1
iterations = 16000
Xm = X[5,:,:]
Ym = Y[5,:,:]
Xm = normalize(Xm)
#Ym = normalize(Ym)
accuracies = []
for i in range(iterations):
    a1, a2 = forward_prop(Xm, Ym, w1, w2, b1, b2)
    dw1, dw2, db1, db2 = back_prop(Xm, Ym, w1, w2, b1, b2, a1, a2, m)
    w1 = w1 - alpha*dw1
    w2 = w2 - alpha*dw2
    b1 = b1 - db1
    b2 = b2 - db2
    acc, res, pred = accuracy(Xm, Ym, w1, w2, b1, b2)
    accuracies.append(acc)
acc, res, pred = accuracy(Xm, Ym, w1, w2, b1, b2)
print(acc)
```
